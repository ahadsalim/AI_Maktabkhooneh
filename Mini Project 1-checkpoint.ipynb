## Importing the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load and Prepare Data

```python
gp_data = pd.read_csv("googleplaystore.csv")
gp_data.head()
```

```python
gp_data.columns = [c.replace(' ', '_') for c in gp_data.columns]
```

```python
gp_data.info()
```

## EDA

```python
gp_data['Type'].value_counts()
```

```python
gp_data['Genres'].value_counts(normalize=True)*100
```

```python
pd.crosstab(gp_data.Rating , gp_data.Price)
```

## Data Preprocessing

```python
gp_data[(gp_data.Rating >= 4.0 ) & ( gp_data.Reviews >100 ) ][['App','Category','Installs']]
```

```python
gp_data[ gp_data.Reviews > 100 ]["Rating"].value_counts()
```

```python
gp_data.groupby('Type')['Rating'].mean()
```

## Strorytelling - Visualization

```python
sns.displot(gp_data, x="Rating",hue="Type", kind="kde" , fill="True")
```

```python
sns.catplot(x="Category", y="Installs", hue="Type" , data=gp_data)
```

```python
g = sns.relplot(x="Rating", y="Installs", kind="line", data=gp_data)
```
